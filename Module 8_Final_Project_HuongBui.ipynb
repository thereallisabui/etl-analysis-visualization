### 2.1 The ETL Process

The ETL (Extract, Transform, Load) process for the MRTS dataset is a structured and systematic approach to prepare the data for integration into a MySQL database and further analysis.

1. Extract:
This phase involves sourcing the dataset and preparing it for subsequent steps. We focus on removing irrelevant and unstructured rows, consolidating data from multiple sources, and adding a 'year' column for segmentation. The data is then exported as a CSV file for further processing.

2. Transform:
The transformation phase aims to make the data analysis-ready. We import the CSV file into a Pandas dataframe, perform data cleaning to remove redundancy and handle missing values, and reshape the data to align with analytical needs.

3. Load:
In the loading phase, we integrate the cleaned and reshaped data into a MySQL database. We create a dedicated MySQL table ('mrts') for structured data storage and insert the processed data into the respective fields. This step ensures that the data is ready for efficient analysis and reporting.

The ETL process ensures that the MRTS dataset is optimally prepared and structured for effective use in analytics and database integration.

### 3.2 Running Queries From Python

To test the previous SQL queries on the MRTS dataset using a Python script and the Terminal window, I followed these steps:

- Python Script Setup: I created a Python script to execute the SQL query. The script included code to establish a connection with the MySQL database, execute the query, and retrieve the results.

- Import Necessary Libraries: I imported the required libraries, including mysql.connector for MySQL database connectivity and matplotlib to visualize data resulted from the query.

- Database Connection: I configured the database connection details such as host, username, password, and database name within the Python script.

- SQL Query Execution: I included the SQL query within the Python script, exactly as it was written, to calculate the sum of sales values for the specified kind of business by year.

- Execution and Result Retrieval: I ran the Python script in the Terminal window. The script established a connection with the MySQL database, executed the SQL query, and retrieved the results.

- Validation: I reviewed the query results displayed in the Terminal to verify that the sum of sales values for the specified kind of business by year was calculated correctly.

By following these steps, I was able to effectively test the SQL query on the MRTS dataset using a Python script in the Terminal window.

```python
# Import necessary libraries
import mysql.connector
import matplotlib.pyplot as plt

# Database connection parameters
host = 'localhost'
user = 'root'
password = 'MyPassword'
database = 'mrtssales'

# Establish a connection to the MySQL database using provided parameters
cnx = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database
)

# Create a cursor object to interact with the database
cursor = cnx.cursor()

# Define the SQL query to retrieve data from the database
sql_retail = ("""
    SELECT year, SUM(sales_value) AS sum_sales_value
    FROM mrts
    WHERE kind_of_business ='Retail and food services sales, total'
    GROUP BY 1
""")
              
sql_gasoline = ("""
    SELECT year, SUM(sales_value) AS sum_sales_value
    FROM mrts
    WHERE kind_of_business ='Gasoline stations'
    GROUP BY 1
""")

# Execute the SQL query using the cursor for Retail and Food Services Sales
cursor.execute(sql_retail)

# Initialize empty lists to store data for Retail and Food Services Sales
year_retail = []
sum_sales_value_retail = []

# Iterate through the result set and fetch data for Retail and Food Services Sales
for row in cursor.fetchall():
    year_retail.append(row[0])
    sum_sales_value_retail.append(row[1])

# Close the cursor for Retail and Food Services Sales
cursor.close()

# Execute the SQL query using the cursor for Gasoline Station Sales
cursor = cnx.cursor()
cursor.execute(sql_gasoline)

# Initialize empty lists to store data for Gasoline Station Sales
year_gasoline = []
sum_sales_value_gasoline = []

# Iterate through the result set and fetch data for Gasoline Station Sales
for row in cursor.fetchall():
    year_gasoline.append(row[0])
    sum_sales_value_gasoline.append(row[1])

# Close the cursor and the database connection
cursor.close()
cnx.close()

# Create line plots using Matplotlib
plt.figure(figsize=(10, 5))

# Plot for Retail and Food Services Sales
plt.subplot(2, 1, 1)
plt.plot(year_retail, sum_sales_value_retail)
plt.title('Retail and Food Services Sales (Total)')
plt.xlabel('Year')
plt.ylabel('Total Sales Value')
plt.grid(True)
plt.ticklabel_format(style='plain', axis='y')

# Plot for Gasoline Station Sales
plt.subplot(2, 1, 2)
plt.plot(year_gasoline, sum_sales_value_gasoline)
plt.title('Gasoline Station Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales Value')
plt.grid(True)
plt.ticklabel_format(style='plain', axis='y')

# Adjust spacing between plots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Establish a connection to the MySQL database
cnx = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database
)

# Create a cursor to interact with the database
cursor = cnx.cursor()

# Define SQL queries for various retail and food services categories
sql_1 = ("""
    SELECT year, SUM(sales_value) AS sum_sales_value
    FROM mrts
    WHERE kind_of_business = 'Retail and food services sales, total'
    GROUP BY 1
    ORDER BY 1;
""")
         
sql_2 = ("""
    SELECT year, SUM(sales_value) AS sum_sales_value
    FROM mrts
    WHERE kind_of_business = 'Retail sales and food services excl motor vehicle and parts'
    GROUP BY 1
    ORDER BY 1;
""")

sql_3 = ("""
    SELECT year, SUM(sales_value) AS sum_sales_value
    FROM mrts
    WHERE kind_of_business = 'Retail sales and food services excl gasoline stations'
    GROUP BY 1
    ORDER BY 1;
""")       

sql_4 = ("""
    SELECT year, SUM(sales_value) AS sum_sales_value
    FROM mrts
    WHERE kind_of_business = 'Retail sales and food services excl motor vehicle and parts and gasoline stations'
    GROUP BY 1
    ORDER BY 1;
""")  

# Execute the SQL query for the first category
cursor.execute(sql_1)

# Initialize lists to store data for the first category
year_1 = []
sum_sales_value_1 = []

# Fetch and store the results for the first category
for row in cursor.fetchall():
    year_1.append(row[0])
    sum_sales_value_1.append(row[1])

# Create a new cursor for the second category query (optional, can reuse previous cursor)
cursor = cnx.cursor()

# Execute the SQL query for the second category
cursor.execute(sql_2)

# Initialize lists to store data for the second category
year_2 = []
sum_sales_value_2 = []

# Fetch and store the results for the second category
for row in cursor.fetchall():
    year_2.append(row[0])
    sum_sales_value_2.append(row[1])

# Create a new cursor for the third category query (optional, can reuse previous cursor)
cursor = cnx.cursor()

# Execute the SQL query for the third category
cursor.execute(sql_3)

# Initialize lists to store data for the third category
year_3 = []
sum_sales_value_3 = []

# Fetch and store the results for the third category
for row in cursor.fetchall():
    year_3.append(row[0])
    sum_sales_value_3.append(row[1])

# Create a new cursor for the fourth category query (optional, can reuse previous cursor)
cursor = cnx.cursor()

# Execute the SQL query for the fourth category
cursor.execute(sql_4)

# Initialize lists to store data for the fourth category
year_4 = []
sum_sales_value_4 = []

# Fetch and store the results for the fourth category
for row in cursor.fetchall():
    year_4.append(row[0])
    sum_sales_value_4.append(row[1])

# Create a figure for the plot
plt.figure(figsize=(10,6))

# Plot the yearly sales trends for each category
plt.plot(year_1, sum_sales_value_1, label="Retail and food services sales, total")
plt.plot(year_2, sum_sales_value_2, label="Retail sales and food services excl motor vehicle and parts")
plt.plot(year_3, sum_sales_value_3, label="Retail sales and food services excl gasoline stations")
plt.plot(year_4, sum_sales_value_4, label="Retail sales and food services excl motor vehicle and parts and gasoline stations")

# Set the y-axis labels to plain style
plt.ticklabel_format(style='plain', axis='y')

# Display the legend for clarity
plt.legend()

# Set labels and title for the plot
plt.xlabel("Year")
plt.ylabel("Total Sales Value")
plt.title("Yearly Sales trends for Retail and food services categories")

# Display grid lines
plt.grid(True)

# Ensure tight layout for better presentation
plt.tight_layout()

# Show the plot
plt.show()
```

### 3.4 Explore Percentage Change

**Percentage Change in Economics:**

Percentage change is a crucial measure in economics as it helps to understand how a specific quantity or value has evolved over time. It is calculated as the ((New Value - Old Value) / Old Value) * 100%. This measure is important because it provides insights into the growth or decline of economic variables like spending patterns. By analyzing percentage changes in various economic indicators, such as consumer spending, GDP, or business revenues, economists can identify trends and make informed predictions about future economic conditions.

Comparison of Women's and Men's Clothing Businesses:

**QUERY 2:** 

Percentage Change of Total Sales for both women's and men's clothing businesses over multiple years

```python
# Establish a connection to the MySQL database
cnx = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database
)

cursor = cnx.cursor()

# Fetch data for women's clothing stores
sql_women = ("""
    SELECT year, SUM(sales_value) AS sum_sales_value_women
    FROM mrts
    WHERE kind_of_business = "Women's clothing stores"
    GROUP BY 1
    ORDER BY 1;
""")

cursor.execute(sql_women)

# Fetch data into a DataFrame for women's clothing stores
data_women = cursor.fetchall()
df_women = pd.DataFrame(data_women, columns=['year', 'sum_sales_value_women'])

# Calculate percentage changes for women's clothing stores using pct_change()
df_women['percentage_change_women'] = df_women['sum_sales_value_women'].pct_change() * 100

# Fetch data for men's clothing stores
sql_men = ("""
    SELECT year, SUM(sales_value) AS sum_sales_value_men
    FROM mrts
    WHERE kind_of_business = "Men's clothing stores"
    GROUP BY 1
    ORDER BY 1;
""")

cursor.execute(sql_men)

# Fetch data into a DataFrame for men's clothing stores
data_men = cursor.fetchall()
df_men = pd.DataFrame(data_men, columns=['year', 'sum_sales_value_men'])

# Calculate percentage changes for men's clothing stores using pct_change()
df_men['percentage_change_men'] = df_men['sum_sales_value_men'].pct_change() * 100

cursor.close()
cnx.close()

# Plot the data
plt.figure(figsize=(10,6))
plt.plot(df_women['year'], df_women['percentage_change_women'], label="Women's Clothing Stores")
plt.plot(df_men['year'], df_men['percentage_change_men'], label="Men's Clothing Stores")
plt.xlabel('Year')
plt.ylabel('Percentage Change (%)')
plt.title('Percentage Change in Sales Over Time')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

This plot illustrates a decreasing pattern in sales for Men's clothing stores throughout the year. Women's clothing sales exhibit a relatively consistent trend, but overall showing a slight negative trend (< 0% percentage change), possibly reflecting the growing influence of online shopping. As mentioned earlier, the year 2020 saw a notable sales decline in both categories, leading to a significant drop in percentage change.

Contribution percentage to Total Sales

```python
# Calculate the total sales combining women's and men's clothing
df_total = df_women['sum_sales_value_women'] + df_men['sum_sales_value_men']

# Calculate the percentage contribution of women's and men's clothing to the total sales
df_women['percentage_change_contribution_women'] = (df_women['sum_sales_value_women'] / df_total) * 100
df_men['percentage_change_contribution_men'] = (df_men['sum_sales_value_men'] / df_total) * 100

# Create a figure for the plot
plt.figure(figsize=(10,6))

# Plot the percentage contribution of women's clothing
plt.plot(df_women['year'], df_women['percentage_change_contribution_women'], label="Women's Clothing")

# Plot the percentage contribution of men's clothing
plt.plot(df_women['year'], df_men['percentage_change_contribution_men'], label="Men's Clothing")

# Set labels and title for the plot
plt.xlabel('Year')
plt.ylabel('Percentage Contribution (%)')
plt.title('Percentage Contribution of Women\'s and Men\'s Clothing to Total Sales Over Time')

# Display the grid and rotate x-axis labels for better readability
plt.grid(True)
plt.xticks(rotation=45)

# Show legend for clarity
plt.legend()

# Display the plot
plt.show()
```

### 3.5 Explore Rolling Time Windows

In economics, a rolling time window, also known as a moving average, is a statistical technique applied to time-series data. It involves calculating the average of a set number of consecutive data points within a sliding window. As the window moves through the data, older data points exit, and newer ones enter, allowing for continuous analysis.

Rolling time windows are a vital tool for predicting quantities like spending patterns for several reasons. 

- They smooth out short-term fluctuations and noise in data, making it easier to identify underlying trends. By reducing the impact of outliers or random fluctuations, rolling averages provide a clearer picture of the overall direction in which spending is moving over time.

- They aid in detecting seasonal patterns. For instance, retailers can use rolling averages to identify increased spending during holiday seasons or specific months of the year, enabling them to tailor marketing and inventory strategies accordingly.

- Valuable for forecasting future spending patterns. By analyzing the trend established through rolling averages, businesses and policymakers can make more accurate predictions and plan accordingly. It also helps in identifying turning points, such as shifts from reduced spending to increased consumer activity.

- Instrumental in risk management, as they provide insights into economic trends, helping organizations prepare for economic downturns or capitalize on periods of growth. Furthermore, they assist in anomaly detection, allowing for the timely response to unexpected changes in spending behavior.

In summary, rolling time windows enhance the analysis and prediction of spending patterns in economics by offering a smoothed, trend-focused perspective on time-series data. This aids in decision-making processes for businesses, policymakers, and economists.

Following is the code to analyze and produce graphs of rolling time windows for the 2 chosen categories, specifically **'New car dealers'** and **'Used car dealers'**:

```python
# Establish a connection to the MySQL database
cnx = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database
)

cursor = cnx.cursor()

# SQL queries for new car and used car dealers
sql_new_car = """
    SELECT date, SUM(sales_value) AS sum_sales_value_new_car
    FROM mrts
    WHERE kind_of_business = 'New car dealers'
    GROUP BY date
    ORDER BY date;
"""

sql_used_car = """
    SELECT date, SUM(sales_value) AS sum_sales_value_used_car
    FROM mrts
    WHERE kind_of_business = 'Used car dealers'
    GROUP BY date
    ORDER BY date;
"""

# Execute the SQL queries
cursor.execute(sql_new_car)
data_new_car = cursor.fetchall()

cursor.execute(sql_used_car)
data_used_car = cursor.fetchall()

# Create DataFrames
df_new_car = pd.DataFrame(data_new_car, columns=['date', 'sum_sales_value_new_car'])
df_used_car = pd.DataFrame(data_used_car, columns=['date', 'sum_sales_value_used_car'])

# Convert 'date' columns to datetime and set as index
df_new_car['date'] = pd.to_datetime(df_new_car['date'])
df_new_car.set_index('date', inplace=True)

df_used_car['date'] = pd.to_datetime(df_used_car['date'])
df_used_car.set_index('date', inplace=True)

# Calculate the rolling means
rolling_window_new_car = df_new_car['sum_sales_value_new_car'].rolling(window=3)
rolling_mean_new_car = rolling_window_new_car.mean()

rolling_window_used_car = df_used_car['sum_sales_value_used_car'].rolling(window=3)
rolling_mean_used_car = rolling_window_used_car.mean()

# Plot the rolling means
plt.figure(figsize=(10, 6))
plt.plot(df_new_car.index, rolling_mean_new_car, label="New car dealers")
plt.plot(df_used_car.index, rolling_mean_used_car, label="Used car dealers")
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
plt.xlabel("Date")
plt.ylabel("Total Sales Value")
plt.title("3-Month Rolling Sales Trends")
plt.grid(True)
plt.show()

# Close the cursor and the database connection
cursor.close()
cnx.close()
```
